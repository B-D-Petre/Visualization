```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
from collections import OrderedDict
pd.set_option('display.max_columns', None)
```

```python
# Set the path to the file you'd like to load
file_paths = ["high_popularity_spotify_data.csv", "low_popularity_spotify_data.csv"]
database = []

# Load the latest version
for file_path in file_paths:
    df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "solomonameh/spotify-music-dataset",
    file_path,
    # Provide any additional arguments like 
    # sql_query or pandas_kwargs. See the 
    # documenation for more information:
    # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
    )
    database.append(df)

```

```python
display(database[0].head())
```

```python
# Create a grid of time series line plots for selected features
import plotly.subplots as sp

# List of features to plot
timeseries_features = [
    'energy', 'tempo', 'danceability', 'loudness', 'liveness', 'valence', 'speechiness', 'acousticness'
]

# Normalize the selected features to 0-1 range before plotting
def normalize_column(col):
    return (col - col.min()) / (col.max() - col.min())

# Create a copy to avoid modifying the original data
normalized_df = database[0].copy()
for feature in timeseries_features:
    normalized_df[feature] = normalize_column(normalized_df[feature])

# Create subplots: 2 rows x 4 columns
grid_rows = 2
grid_cols = 4
fig = sp.make_subplots(rows=grid_rows, cols=grid_cols, subplot_titles=timeseries_features)

# Plot each feature as a line plot (mean per year, normalized)
for i, feature in enumerate(timeseries_features):
    row = i // grid_cols + 1
    col = i % grid_cols + 1
    # Group by year and calculate mean
    ts = normalized_df.groupby('track_album_release_date')[feature].mean().reset_index()
    # If 'track_album_release_date' is not datetime, convert it
    if not pd.api.types.is_datetime64_any_dtype(ts['track_album_release_date']):
        ts['track_album_release_date'] = pd.to_datetime(ts['track_album_release_date'], errors='coerce')
    ts = ts.sort_values('track_album_release_date')
    fig.add_trace(
        go.Scatter(x=ts['track_album_release_date'], y=ts[feature], mode='lines', name=feature),
        row=row, col=col
    )

fig.update_layout(
    height=600, width=1200, showlegend=False,
    title_text="Time Series of Normalized Audio Features (Mean per Year)",
    margin=dict(l=20, r=20, t=40, b=20),
    autosize=True
)
fig.update_xaxes(matches=None, showticklabels=True)
fig.update_yaxes(matches=None)
fig.show()
```

# Alternatives to Line Plots for Time Series Audio Features
Below are several alternative visualizations for the normalized audio features over time.

```python
# Area Plots: Area under the curve for each feature
fig_area = sp.make_subplots(rows=2, cols=4, subplot_titles=timeseries_features)
for i, feature in enumerate(timeseries_features):
    row = i // 4 + 1
    col = i % 4 + 1
    ts = normalized_df.groupby('track_album_release_date')[feature].mean().reset_index()
    if not pd.api.types.is_datetime64_any_dtype(ts['track_album_release_date']):
        ts['track_album_release_date'] = pd.to_datetime(ts['track_album_release_date'], errors='coerce')
    ts = ts.sort_values('track_album_release_date')
    fig_area.add_trace(
        go.Scatter(x=ts['track_album_release_date'], y=ts[feature], fill='tozeroy', mode='lines', name=feature),
        row=row, col=col
    )
fig_area.update_layout(height=600, width=1200, showlegend=False, title_text="Area Plots of Normalized Audio Features (Mean per Year)")
fig_area.show()
```

```python
# Bar Plots: Bar chart for each feature per year
fig_bar = sp.make_subplots(rows=2, cols=4, subplot_titles=timeseries_features)
for i, feature in enumerate(timeseries_features):
    row = i // 4 + 1
    col = i % 4 + 1
    ts = normalized_df.groupby('track_album_release_date')[feature].mean().reset_index()
    if not pd.api.types.is_datetime64_any_dtype(ts['track_album_release_date']):
        ts['track_album_release_date'] = pd.to_datetime(ts['track_album_release_date'], errors='coerce')
    ts = ts.sort_values('track_album_release_date')
    fig_bar.add_trace(
        go.Bar(x=ts['track_album_release_date'], y=ts[feature], name=feature),
        row=row, col=col
    )
fig_bar.update_layout(height=600, width=1200, showlegend=False, title_text="Bar Plots of Normalized Audio Features (Mean per Year)")
fig_bar.show()
```

```python
# Heatmap: Features as rows, years as columns
# Prepare data for heatmap
heatmap_df = normalized_df.copy()
heatmap_df['year'] = pd.to_datetime(heatmap_df['track_album_release_date'], errors='coerce').dt.year
heatmap_matrix = heatmap_df.groupby('year')[timeseries_features].mean().T
fig_heatmap = go.Figure(data=go.Heatmap(
    z=heatmap_matrix.values,
    x=heatmap_matrix.columns,
    y=heatmap_matrix.index,
    colorscale='Viridis',
    colorbar=dict(title='Normalized Value')
))
fig_heatmap.update_layout(height=400, width=900, title="Heatmap of Normalized Audio Features by Year", xaxis_title="Year", yaxis_title="Feature")
fig_heatmap.show()
```

```python
# Box Plots: Distribution of each feature per year
fig_box = sp.make_subplots(rows=2, cols=4, subplot_titles=timeseries_features)
for i, feature in enumerate(timeseries_features):
    row = i // 4 + 1
    col = i % 4 + 1
    temp_df = normalized_df.copy()
    temp_df['year'] = pd.to_datetime(temp_df['track_album_release_date'], errors='coerce').dt.year
    fig_box.add_trace(
        go.Box(x=temp_df['year'], y=temp_df[feature], name=feature, boxmean=True, marker_color='royalblue'),
        row=row, col=col
    )
fig_box.update_layout(height=600, width=1200, showlegend=False, title_text="Box Plots of Normalized Audio Features by Year")
fig_box.show()
```

```python
# Violin Plots: Distribution shape of each feature per year
fig_violin = sp.make_subplots(rows=2, cols=4, subplot_titles=timeseries_features)
for i, feature in enumerate(timeseries_features):
    row = i // 4 + 1
    col = i % 4 + 1
    temp_df = normalized_df.copy()
    temp_df['year'] = pd.to_datetime(temp_df['track_album_release_date'], errors='coerce').dt.year
    fig_violin.add_trace(
        go.Violin(x=temp_df['year'], y=temp_df[feature], name=feature, box_visible=True, meanline_visible=True, line_color='purple'),
        row=row, col=col
    )
fig_violin.update_layout(height=600, width=1200, showlegend=False, title_text="Violin Plots of Normalized Audio Features by Year")
fig_violin.show()
```

```python
# Scatter Plots: Mean value per year for each feature
fig_scatter = sp.make_subplots(rows=2, cols=4, subplot_titles=timeseries_features)
for i, feature in enumerate(timeseries_features):
    row = i // 4 + 1
    col = i % 4 + 1
    ts = normalized_df.groupby('track_album_release_date')[feature].mean().reset_index()
    if not pd.api.types.is_datetime64_any_dtype(ts['track_album_release_date']):
        ts['track_album_release_date'] = pd.to_datetime(ts['track_album_release_date'], errors='coerce')
    ts = ts.sort_values('track_album_release_date')
    fig_scatter.add_trace(
        go.Scatter(x=ts['track_album_release_date'], y=ts[feature], mode='markers', name=feature),
        row=row, col=col
    )
fig_scatter.update_layout(height=600, width=1200, showlegend=False, title_text="Scatter Plots of Normalized Audio Features (Mean per Year)")
fig_scatter.show()
```
